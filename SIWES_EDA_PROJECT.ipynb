```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

```python
trained=pd.read_csv("train.csv",parse_dates=['date_id'])
```

```python
trained.set_index('date_id', inplace=True)
```

```python
missing = trained.isnull().sum()
print(missing[missing > 0])
```

```python
# Visualize missing values
plt.figure(figsize=(12, 6))
sns.heatmap(trained.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

visualize time series trend

```python
 #Example: LME metal closes
lme_cols = [col for col in trained.columns if 'LME' in col and 'Close' in col]
trained[lme_cols].plot(figsize=(20, 8), title='LME Metal Closing Prices')
plt.show()
```

```python
# Example: JPX Gold Futures Open
jpx_gold_cols = [col for col in trained.columns if 'JPX_Gold' in col and 'Open' in col]
trained[jpx_gold_cols].plot(figsize=(20, 8), title='JPX Gold Futures Open Prices')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Example: LME Aluminum Close
series = trained['LME_AH_Close'].dropna()
result = seasonal_decompose(series, model='additive', period=30)  # adjust period as needed
result.plot()
plt.show()
```

```python
trained.plot(subplots = True, figsize = (10,12))
plt.title('NVidia Stock Attributes')
plt.show()
```
